**Margin:** Gap between hyperplan(φ) and the point

**Margin maximizing hyperplane: **In SVM the hyperplan(φ) will act as a Margin maximizing hyperplane means it will also maximizes the margin when it classifies the data so that the resultant model is generalized. 

**Steps:**
1. Randomly selects a hyperplane.
2. Draw more hyperplanes by going along the parallel direction of that plane untill you reach the first green point. Call this next finded hyperplane as pi psoitive(φ+)
3. Same like this, find another hyperplane oppoeite to pi(φ). Call this  hyperplane as pi negative(φ-)
4. Calculate the margin d (the shortest distance between φ+ and φ-)
5. Decide one particular plane for which margin value is maximum.
