# **Updating values in DataFrame**

Tutorial link: https://youtu.be/DCDe29sIKcE

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.columns
```

```python
df.columns = ['first_name', 'last_name', 'email'] #renaming columns
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
df
```

```python
df.loc[2] = ['John', 'Smith', 'JohnSmith@email.com']
df
```

```python
df.loc[2, ['last', 'email']] = ['Doe', 'JohnDoe@email.com']
df
```

```python
df.loc[2, 'last'] = 'Smith'
df
```

```python
df.at[2, 'last'] = 'Doe' #alternate
df
```

```python
filt = (df['email'] == 'JohnDoe@email.com')
df[filt]['last'] = 'Smith'
```

```python
filt = (df['email'] == 'JohnDoe@email.com')
df.loc[filt, 'last'] = 'Smith'
df
```

```python
df['email'] = df['email'].str.lower()
df
```

```python
df['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()

df['email'].apply(update_email)
```

```python
df['email'] = df['email'].apply(update_email)
df
```

```python
df['email'] = df['email'].apply(lambda x: x.lower())
df
```

- lambda  is a way to define a function in a single line of code without giving it a name.
- Here lambda function is used to apply the lower() method to each element in the df['email'] column of the Pandas DataFrame.
