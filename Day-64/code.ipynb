**Manual Feature Extraction:**

1. Utilizing domain knowledge.
2. Defining specific features.
3. Detecting features to enable classification.

Due to the difficulty of manually detecting features, we can employ neural networks that autonomously learn and extract features.

**Spatial structure in CNN:**

It refers to the arrangement and relationships of pixels in an image. In simple terms, CNNs work by calculating weights that are applied to different parts of the image. These *weights capture important patterns and features*, such as edges, textures, and shapes. By convolving these weights with the input image, the CNN can detect and emphasize relevant features while disregarding irrelevant information.

By incorporating this spatial understanding, CNNs excel in tasks like image recognition, object detection, and image segmentation.

**Feature Extraction in Image Classification:**

Feature extraction is the process of identifying and capturing meaningful patterns or features from raw input data, such as images. In image classification, it involves automatically identifying important characteristics that can distinguish different classes of images.

**Convolution Operation in Image Classification:**

The convolution operation acts like a spotlight, focusing on different parts of the image at a time. By applying multiple filters, CNNs can detect various features such as edges, textures, and shapes. These extracted features are then used for classification tasks, helping the model make informed decisions about the content of the image.
